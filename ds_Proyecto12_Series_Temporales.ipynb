## Análisis

A continuación, realizaremos un análisis de los datos para poder observar su tendencia y estacionalidad. Para esto realizaremos un remuestreo por una hora, y analizaremos cambios tanto de marzo a agosto, los cambios en un mes y en un día específico de nuestro dataset. 

### Número de pedidos por hora Marzo-Agosto 2018

```python
# Realizamos el remuestreo por una hora y calculamos la media y desviación estándar a través de rolling
data = data.resample('1H').sum()
mean = data['num_orders'].rolling(15).mean()
std = data['num_orders'].rolling(15).std()

# Trazamos un gráfico de líneas
data.plot(figsize=(12,6), title='Número de pedidos por hora de Marzo-Agosto de 2018')
mean.plot()
std.plot()
plt.legend(['num_orders', 'mean', 'std'])
plt.show()
```

Al realizar un remuestreo por hora, podemos observar un promedio que incrementa a partir de Julio y una desviación estándar con fluctuaciones a lo largo de los seis meses, lo que correspondría a un proceso estocástico no estacionario. Ahora vamos a revisar la tendencia y estacionalidad de nuestros datos por hora.
 

```python
# Llamamos a la función seasonal_decompose para obtener la tendencia, estacionalidad y residuos de nuestros datos x hora
decomposed_h = seasonal_decompose(data)

# Fijamos el tamaño y el título de nuestro gráfico que contendrá tres subplots
plt.figure(figsize=(12, 10))
plt.suptitle('Tendencia, estacionalidad y residuos por hora Marzo-Agosto 2018', fontsize=14)

# plt.subplot(x,y,z) dice que las imágenes forman una tabla de x filas y y columnas, z es el lugar de la imagen actual.
plt.subplot(311) # plt.subplot(311) dice 3 filas, 1 columna, primera imagen. 
decomposed_h.trend.plot(ax=plt.gca())
plt.title('Tendencia')
plt.subplot(312) # plt.subplot(312) dice 3 filas, 1 columna, segunda imagen.
decomposed_h.seasonal.plot(ax=plt.gca())
plt.title('Estacionalidad')
plt.subplot(313) # plt.subplot(313) dice 3 filas, 1 columna, tercera imagen.
decomposed_h.resid.plot(ax=plt.gca())
plt.title('Residuales')

# Encajamos gráficos secundarios en el área 
plt.tight_layout()
plt.show()
```

La tendencia se define como los cambios ligeros del valor medio de una serie temporal sin repetir patrones, en el caso del número de pedidos de taxis se identifica claramente una tendencia al incremento del número de pedidos entre el período de Marzo a Agosto de 2018. Con un incremento por encima de los 2000 pedidos a partir de Junio, lo que estaría relacionado con el inicio de verano y por tanto la llegada de vacaciones, haciendo que incrementen los pedidos de taxis en aeropuertos. En el caso de la estacionalidad no podemos observar un patrón cíclico al trabajar con nuestros datos por hora, así que realizaremos un remuestreo por día para analizar nuestros datos. 

### Número de pedidos por día Marzo-Agosto

Realizamos un remuestreo por día y luego trazamos un gráfico de líneas para visualizar los resultados.

```python
# Remuestreo por día 
data_day = data.resample('1D').sum()

# Trazamos gráfico para observar resultados
data_day.plot(figsize=(12,6), title='Número de pedidos por día de Marzo-Agosto 2018')
plt.show()
```

Se observa un incremento en el número de pedidos a partir de Junio 2018, alcanzando valores por encima de 3000 pedidos a partir de Agosto de 2018. Al realizar un remuestreo por día se observan más fácilmente cómo se comportan nuestros datos, pudiendo corroborar que un cambio en la escala nos permite una mejor identificación de patrones. También hay que resaltar la presencia de varios picos a lo largo de este período de tiempo, por lo que será necesario analizar más a fondo este comportamiento.

Ahora vamos a llamar a la función `seasonal_decompose` de la librería `StatsModels` para visualizar la tendencia, estacionalidad y residuos por día de Marzo-Agosto 2018.

```python
# Llamamos a la función seasonal_decompose para obtener la tendencia, estacionalidad y residuos de nuestros datos por día
decomposed = seasonal_decompose(data_day)

# Fijamos el tamaño y el título de nuestro gráfico que contendrá tres subplots
plt.figure(figsize=(12, 10))
plt.suptitle('Tendencia, estacionalidad y residuos por día Marzo-Agosto 2018', fontsize=14)

# plt.subplot(x,y,z) dice que las imágenes forman una tabla de x filas y y columnas, z es el lugar de la imagen actual.
plt.subplot(311) # plt.subplot(311) dice 3 filas, 1 columna, primera imagen. 
decomposed.trend.plot(ax=plt.gca())
plt.title('Tendencia')
plt.subplot(312) # plt.subplot(312) dice 3 filas, 1 columna, segunda imagen.
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Estacionalidad')
plt.subplot(313) # plt.subplot(313) dice 3 filas, 1 columna, tercera imagen.
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuales')

# Encajamos gráficos secundarios en el área 
plt.tight_layout()
plt.show()
```

En la tendencia por día se observa un incremento por encima de los 2000 pedidos a partir de Junio, lo que estaría relacionado con el inicio de verano y por tanto la llegada de vacaciones, haciendo que incrementen los pedidos de taxis en aeropuertos. 

En cuanto a la estacionalidad, que se define como aquellos patrones que se repiten de forma cíclica en una serie temporal, se evidencian ocho picos en cada mes y cada par de picos se repite cada semana, es decir que existe un patrón de incremento en el pedido de taxis dos veces por semana. Vamos a analizar posteriormente uno de los meses de estudio para encontrar a qué se debe este patrón cíclico en nuestros datos. 

Establecido que la tendencia está más relacionada con el inicio de verano, mientras que la estacionalidad es un patrón semanal, vamos a trazar la desviación estándar y media móvil de nuestro datos remuestreados por día. 

```python
# Calculamos la media y desviación estándar a partir de la media móvil
data_day['mean'] = data_day['num_orders'].rolling(7).mean()
data_day['std'] = data_day['num_orders'].rolling(7).std()

# Trazamos un gráfico de líneas
data_day.plot(figsize=(12,6), title='Desviación Estándar y Media Móvil: número de pedidos por día Marzo-Agosto 2018')
plt.show()
```

Nuestros datos cumplen con los patrones de un proceso estocásico no estacionario, es decir una serie temporal con variación aleatoria y con una media y desviación estándar que cambian a lo largo del tiempo. Podemos establecer un incremento de la media, mientras que la desviación estándar presenta pequeñas fluctuaciones pero se mantiene alrededor del mismo valor a lo largo del tiempo. 

Anteriormente observamos ocho picos que correspondían a patrones mensuales, vamos a analizar el comportamiento de nuestros datos en un mes dado y veremos a qué se relaciona este patrón. Escogeremos el mes de Junio ya que es el mes en el cual inicia verano y se observa el inicio en el incremento de pedidos de taxis, utilizaremos nuestros datos por día a los cuales ya se aplicó la función `seasonal_decompose` y haremos un segmento en '2018-06'.

### Prueba de estacionariedad

Al observar la media móvil y desviación estándar de nuestros datos remuestreados por hora y día, pudimos observar una tendencia al incremento del número de pedidos por lo que establecimos que nuestro dataset correspondía a un proceso estocástico no estacionario. Sin embargo, no podemos confirmar esto sin una prueba estadística. En este caso utilizaremos la prueba de Dickey-Fuller aumentada, esta es una prueba de raíz única que detecta estadísticamente la presencia de tendencias, ya sea estacionarias o no estacionarias, en series temporales mediante un contraste de hipótesis. Esta prueba se fundamenta en las siguientes hipotesis:

- H0 = la serie temporal no es estacionaria o presenta una raíz unitaria
- H1 = la serie temporal es estacionaria y no presenta una raíz unitaria

Utilizaremos un nivel de significancia del 5%, si el valor p es mayor a 0.05 aceptaremos nuestra hipótesis nula y la serie temporal no es estacionaria. A su vez, si el valor p es menor a 0.05 rechazaremos la hipótesis nula y nuestra series es estacionaria. 

```python
# Pasamos los valores del número de pedidos a la función adfuller y lo guardamos en la variable res
res = adfuller(data['num_orders'])
 
# Imprimimos el valor estadístico de la prueba Dickey-fuller
print('Valor del Estadístico Dickey-fuller: %f' % res[0])
print('Valor p: %f' % res[1])

# Imprimimos los valores críticos a diferentes valores alpha
print('Valores críticos a diferentes niveles de significancia:')
for k, v in res[4].items():
    print('\t%s: %.3f' % (k, v))
```

El valor p obtenido 0.028 inferior a nuestro nivel de significancia de 0.05, además el valor del estadístico Dickey-fuller -3.0689 es inferior al valor crítico a un nivel del 5%, por lo que rechazamos la hipótesis nula y nuestra serie temporal es en realidad estacionaria. A pesar de que se observó un cierto incremento en el número de pedidos, este no fue significativo como para determinar un proceso estocástico no estacionario. Por lo tanto, no será necesario transformar nuestros datos en una serie temporal estacionaria.

### Creación características

Ya que estamos trabajando con una serie temporal será necesario crear las características que permitirán el entrenamiento de los diferentes modelos. En este caso vamos a crear una función `make_features` que añadirá características del calendario relacionadas a la tendencia y estacionalidad observadas anteriormente, en este caso observamos patrones relacionados con cada mes, día y hora. A su vez, se añadirán características de desfase que permiten utilizar valores anteriores para establecer si el número de pedidos disminuirá o aumentará en el futuro. Finalmente, se utilizará la media móvil para tener una referencia de la tendencia general de nuestros datos. 

```python
def make_features(data, max_lag, rolling_mean_size):
    
    '''
    Devuelve un nuevo dataset con características de calendario, características de desfase y media móvil
    
    :data: nuestra serie temporal
    :max_lag: número máximo de características de desfase que se utilizará
    :rolling_mean_size: tamaño de la media móvil
    
    '''
    # Creamos características de calendario aplicando atributos a las fechas en formato datetime
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day_of_month'] = data.index.day
    data['day_of_week'] = data.index.dayofweek
    data['hour'] = data.index.hour
    
    # Creamos un bucle for que nos devolverá las características de desfase en base a un tamaño de ventana
    for lag in range(1, max_lag + 1):
        data['lag{}'.format(lag)] = data['num_orders'].shift(lag)
    
    # Establecemos la media móvil excluyendo el valor actual
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    return data
```

```python
# Aplicamos nuestra función make_features con 7 características de desfase y 6 como tamaño de ventana
make_features(data, 7, 6)

# Comprobamos cambios
data.head()
```

## Entrenamiento y prueba modelos

Establecidos nuestros conjuntos de entrenamiento y prueba, vamos a entrenar diferentes modelos con diferentes hiperparámetros y medir el rendimiento de cada uno de estos a través de la métrica RECM. Escogeremos aquel modelo con el RECM inferior a 48. Vamos a trabajar con modelos tradicionales de regresión: bosque aleatorio, árbol de decisión y regresión lineal, y con modelos de potenciación de gradiente: LightGBM, CatBoost y XGBoost. 

Iniciamos con la creación de la función `grid_search_model` que nos ayudará a obtener los mejores hiperparámetros que devuelvan el RECM más bajo posible a través de GridSearchCV. 

```python
def grid_search_model(estimator_name, params, x_train, y_train):
    
    '''
    Devuelve los mejores hiperparámetros y el RECM más bajo para un determinado algoritmo de aprendizaje.
    
    : estimator_name: nombre del algoritmo
    : params: hiperparámetros de cada modelo
    : x_train: características entrenamiento
    : y_train: objetivo entrenamiento.
    
    '''
    gridSearch= GridSearchCV(
    estimator=estimator_name,
    param_grid=params,
    scoring='neg_root_mean_squared_error',
    cv=tscv)

    gridSearch.fit(x_train, y_train)

    max_score = gridSearch.cv_results_['mean_test_score'].max()
    index_max_score = np.where(gridSearch.cv_results_['mean_test_score'] == max_score)[0][0]

    best_set_of_params = gridSearch.cv_results_['params'][index_max_score]
    
    return best_set_of_params, np.abs(max_score)
```

Entrenaremos nuestros modelos y obtendremos su rendimiento a través de validación cruzada, ya que estamos trabajando con una serie temporal no podremos realizar una división al azar del conjunto de entrenamiento y validación. Por lo tanto, la mejor opción será definir un `TimeSeriesSplit` que dividirá nuestro dataset de entrenamiento de forma continua, empezará entrenando un segmento y validando el segundo, luego entrenará los dos segmentos anteriores y validará en tercer segmento con nuevas observaciones, de esta manera se evitará mezclar observaciones que son dependientes del tiempo. 

```python
tscv = TimeSeriesSplit(n_splits=5)
```

### Prophet Facebook

Requiere un preprocesamiento diferente a los modelos previamente utilizados, así que empezaremos estableciendo un nuevo dataset que contenga una columna con la fecha `ds` en formato datetime y otra con la variable numérica `y` que se busca predecir. Se utilizaron las siguiente fuentes bibliográficas para desarrollar este modelo:

- [Análisis y Predicción de Series Temporales con FB Prophet Python](https://www.modeldifferently.com/2022/04/analisis_prediccion_ts_prophet/)
- [Time Series forecasting with Prophet](https://www.kaggle.com/code/robikscube/time-series-forecasting-with-prophet)



```python
# Filtramos nuestro dataset original por la columna `num_orders` y aplicamos `reset_index` 
prophet_data = data['num_orders'].reset_index()

# Cambiamos el nombre de las columnas 
prophet_data.columns = ['ds', 'y']

# Imprimimos los resultados
prophet_data.head()
```

Establecemos una muestra para entrenamiento y otra de prueba, manteniendo el 10% de datos iniciales en el conjunto de prueba. Posteriormente, comprobamos que la división se haya realizado correctamente y  verificamos el tamaño de nuestros nuevos datasets.

Comprobado el tamaño de nuestro conjunto de entrenamiento y prueba vamos a realizar el ajuste de hiperparámetros, utilizando diferentes módulos de la librería Prophet. 

**Ajuste Hiperparámetros**

Estableceremos los hiperparámetros que se ajustarán y realizaremos una validación cruzada para obtener la mejor métrica RECM. En esta caso vamos a ajustar dos hiperparámetros:

- `changepoint_prior_scale`: representa la fuerza o capacidad del cambio de tendencia. Cuanto más grande es, más flexibilidad tiene la tendencia, a menor valor, menos flexibilidad.
- `seasonality_prior_scale`: fuerza de la componente o componentes de estacionalidad (en caso de haber varias). Este valor se puede ajustar para cada una de las estacionalidades por separado.
 
Para la validación cruzada establecemos un horizonte de 795 horas, ya que estamos trabajando con valores por hora y este es el valor aproximado de horas en un mes. El horizonte `horizon` se define como el conjunto de observaciones sobre las cuales se realizará las predicciones, es decir el conjunto de validación. Por otro lado, `initial` es el tamaño del conjunto de entrenamiento y `period` es el número de observaciones que se añaden al conjunto de entrenamiento para entrenarlo y realizar las predicciones en un nuevo conjunto de validación. No utilizaremos estos dos últimos argumentos y dejaremos que Prophet escoga los valores óptimos de estos a partir del horizonte. 

Para establecer la métrica RECM utilizaremos `performance_metrics` que devuelve un dataset con los puntajes de difentes métricas, en este caso RECM se encuentra en la primera columna, así que filtraremos estos valores y los añadiremos a la lista rmses. 

```python
# Establecemos los diferentes hiperparámetros que van a ser analizados
param_grid = {  
    'changepoint_prior_scale': [0.001, 0.01, 0.1, 0.5],
    'seasonality_prior_scale': [0.01, 0.1, 1.0, 10.0],   
}

# Generamos todas las combinaciones de hiperparámetros
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
rmses = [] # Almacenamos el RECM para cada conjunto de hiperparámetros en una lista

# Utilizamos validación cruzada para evaluar todos los hiperparámetros
for params in all_params:
    m = Prophet(**params).fit(train_prophet)  # Entrenamos a nuestro conjunto de entrenamiento 
    df_cv = cross_validation(m, horizon = '795 hours', parallel="processes") # Realizamos validación cruzada
    df_p = performance_metrics(df_cv, rolling_window=1) # Obtenemos las métricas de la validación
    rmses.append(df_p['rmse'].values[0]) # Añadimos el RECM a la lista rmses

# Encontramos los mejores hiperparámetros
tuning_results = pd.DataFrame(all_params)
tuning_results['rmses'] = rmses

# Imprimimos los hiperparámetros en orden ascendente en función del valor RECM
tuning_results.sort_values(['rmses'])
```

```python
# Establecemos los mejores hiperparámetros y el mejor valor RECM
best_params_prophet = all_params[np.argmin(rmses)]
best_score_prophet = rmses[np.argmin(rmses)]

print(f'Mejores Hiperparámetros: {best_params_prophet}, Mejor RECM: {best_score_prophet:.3f}')
```

**Entrenamiento modelo Prophet**

Definidos los mejores hiperparámetros vamos a entrenar nuestro modelo en base a estos valores. Añadiremos el hiperparámetro `seasonality_mode` que establece el efecto del resto de variables, es decir si suman o multiplican la tendencia, en este caso utilizaremos multiplicativo. Una vez entrenado el modelo realizaremos predicciones en el conjunto de prueba, Prophet devuelve una tabla con los valores de predicciones ubicados en la columna `yhat`. 

```python
# Aplicamos los hiperparámetros a nuestro modelo 
m = Prophet(seasonality_mode='multiplicative', changepoint_prior_scale=0.5, seasonality_prior_scale=10.0)
# Entrenamos nuestro conjunto de entrenamiento
m.fit(train_prophet)
# Realizamos predicciones en nuestro conjunto de prueba
predict = m.predict(test_prophet)
# Imprimimos las predicciones realizadas
predict.head()
```

```python
# Obtenemos la métrica RECM en base a los valores reales 'test_prophet' y las predicciones 'yhat'
RECM_prophet = mean_squared_error(y_true=test_prophet['y'], y_pred=predict['yhat'])**0.5
print(f'RECM Prophet: {RECM_prophet}')
```
